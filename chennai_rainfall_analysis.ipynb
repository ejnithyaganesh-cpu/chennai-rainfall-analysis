# Chennai Rainfall Data Analysis

## Objective
To analyze rainfall patterns across Tamil Nadu districts using data analysis techniques.

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

## Data Loading

```python
df = pd.read_csv("district wise rainfall normal.csv")
df.head()
```

## Data Exploration

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

## Filtering Tamil Nadu Data

```python
tn = df[df['STATE_UT_NAME'] == 'TAMIL NADU']
tn.head()
```

## Top Districts by Annual Rainfall

```python
top_districts = tn.sort_values(by='ANNUAL', ascending=False).head(10)
top_districts[['DISTRICT', 'ANNUAL']]
```

## Lowest Rainfall Districts

```python
low_districts = tn.nsmallest(10, 'ANNUAL')
low_districts[['DISTRICT', 'ANNUAL']]
```

## Monthly Rainfall Analysis

```python
monthly_avg = tn[['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']].mean()
```

```python
monthly_avg.plot(kind='bar')
plt.title("Average Monthly Rainfall - Tamil Nadu")
plt.xlabel("Month")
plt.ylabel("Rainfall")
plt.xticks(rotation=45)
plt.show()
```

## Seasonal Rainfall Analysis

```python
seasonal = tn[['Jan-Feb','Mar-May','Jun-Sep','Oct-Dec']].mean()
```

```python
seasonal.plot(kind='bar')
plt.title("Seasonal Rainfall Pattern - Tamil Nadu")
plt.xlabel("Season")
plt.ylabel("Rainfall")
plt.show()
```

## Rainfall Distribution

```python
tn['ANNUAL'].plot(kind='hist', bins=20)
plt.title("Distribution of Annual Rainfall")
plt.xlabel("Rainfall")
plt.show()
```

## Top 5 Rainfall Districts Visualization

```python
top5 = tn.nlargest(5, 'ANNUAL')
```

```python
plt.bar(top5['DISTRICT'], top5['ANNUAL'])
plt.title("Top 5 Rainfall Districts")
plt.xlabel("District")
plt.ylabel("Annual Rainfall")
plt.xticks(rotation=45)
plt.show()
```

## Saving Processed Data

```python
tn.to_csv("tamil_nadu_rainfall.csv", index=False)
```

## Key Insights

- The highest rainfall in Tamil Nadu occurs during the Oct–Dec period, indicating the dominance of the Northeast Monsoon.
- Monthly analysis shows peak rainfall in October and November, while summer months (March–May) receive the least rainfall.
- There is significant variation in annual rainfall across districts, highlighting regional differences.
- Certain districts consistently receive higher rainfall compared to others.
- Seasonal analysis confirms that the majority of rainfall is concentrated in specific monsoon periods.
